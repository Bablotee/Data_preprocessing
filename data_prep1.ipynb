### Skew of Univariate Distributions

```python

Skew refers to a distribution that is assumed Gaussian (normal or bell curve) that is shifted or
squashed in one direction or another. 

Many machine learning algorithms assume a Gaussian distribution. 
Knowing that an attribute has a skew may allow you to perform data preparation to correct the skew 
the skew and improve the accuracy of your models. 

You can calculate the skew of each attribute using the skew() function on the Pandas DataFrame.
```

```python
Skew can be positive (right) or negative (left) skew. 
Values closer to zero show less skew.
```

## Understand Your Data With Visualization

### Univariate Plots

```python
three techniques that you can use to understand each attribute of your dataset independently.
 Histograms.
 Density Plots.
 Box and Whisker Plots.
```

```python
A fast way to get an idea of the distribution of each attribute is to look at histograms. Histograms
group data into bins and provide you a count of the number of observations in each bin. 

From the shape of the bins you can quickly get a feeling for whether an attribute is Gaussian, skewed
or even has an exponential distribution. It can also help you see possible outliers.
```

```python
#Univariate Histograms
from matplotlib import pyplot
'''
from pandas import read_csv
filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename)'''


data.hist()
plt.gcf().set_size_inches(20,20)
pyplot.show()
```

```python
the attributes age, pedi and test may have an exponential distribution. 
mass and pres and plas attributes may have a Gaussian or nearly Gaussian distribution. 

many machine learning techniques assume a Gaussian univariate distribution on the input variables
```

### Box and Whisker Plots

```python

Boxplots summarize the distribution of each attribute, 
drawing a line for the median (middle value) and a box around the 25th and 75th 
percentiles (the middle 50% of the data). 

The whiskers give an idea of the spread of the data 
and dots outside of the whiskers show candidate outlier values 
(values that are 1.5 times greater than the size of spread of the middle 50% of the data).
```

```python
# Box and Whisker Plots
from matplotlib import pyplot
from pandas import read_csv
'''filename = "pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename)'''

data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.gcf().set_size_inches(20,20)
pyplot.show()
```

```python
Here we look at two plots that show the interactions between multiple variables in a dataset
. Correlation Matrix Plot.
. Scatter Plot Matrix.
```

```python
Correlation gives an indication of how related the changes are between two variables. 
If two variables change in the same direction they are positively correlated. 

If they change in opposite directions together (one goes up, one goes down), 
then they are negatively correlated. 

You can calculate the correlation between each pair of attributes using a correlation matrix. 

This is useful to know, because some machine learning algorithms like linear
and logistic regression can have poor performance if there are highly correlated input variables
in your data.
```

```python
# Correction Matrix Plot
from matplotlib import pyplot
from pandas import read_csv
import numpy
'''filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename)'''
#names = list(data.columns)

correlations = data.corr()
# plot correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```

```python

A scatter plot shows the relationship between two variables as dots in two dimensions, one
axis for each attribute. 

You can create a scatter plot for each pair of attributes in your data.
Drawing all these scatter plots together is called a scatter plot matrix. 

Scatter plots are useful for spotting structured relationships between 
variables, like whether you could summarize the
relationship between two variables with a line. 

Attributes with structured relationships may also be correlated 
and good candidates for removal from your dataset.
```

```python
the scatter plot matrix is symmetrical. 
This is useful to look at the pairwise relationships from 
different perspectives.  
the diagonal shows histograms of each attribute.
```
